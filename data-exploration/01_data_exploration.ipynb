```python
grouped_data = df.groupby(["fagfelt","studiested"])["indeks_psymiljo_15"].mean()

unstacked_data = grouped_data.unstack()

# The output of unstacked_data looks like this:
# Category2     X     Y
# Category1
# A            11.0  14.5
# B             8.0  19.0

# 4a. Plot the unstacked DataFrame
ax = unstacked_data.plot(kind='bar', figsize=(8, 6), rot=0) # rot=0 prevents x-axis label rotation
plt.title('Mean Value by Category1 and Category2')
plt.ylabel('Mean Value')
plt.xlabel('Category1')
plt.legend(title='Category2')
plt.show()
```
